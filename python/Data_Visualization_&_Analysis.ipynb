# Analyzing Abortion, Pregnancy and Birth Data

```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#using the used datasets
abortion_rate = pd.read_csv("/content/drive/MyDrive/DS4AW Capstone Project-20220911T125049Z-001.zip (Unzipped Files)/DS4AW Capstone Project/abortion_rate (1).csv")
```

```python
abortion_rate.head()
```

```python
abortion_rate.info()
```

```python
abortion_rate.describe()
```

```python
abortion_rate["year"] = abortion_rate["year"].astype("str")
```

```python
df = abortion_rate[(abortion_rate["state"]=="US")]
```

```python
#visualizing abortion rate trends over the years
sns.set_style("darkgrid")
g = sns.relplot(x="year",y="abortionratetotal",kind='line',data=df.sort_values(by=["year"]),ci=None,marker='o',color="r")
plt.xticks(rotation=90)
g.fig.set_size_inches([15,8])
g.fig.suptitle("Trends in Abortion Rate among Women from 1973â€“2017 in United States",x=0.55,fontsize = 16)
g.set(xlabel="Year", 
       ylabel="Abortion Rate per 1000 women")
plt.show()
```

```python
#comparing the various rates over the years
fig, ax = plt.subplots()
fig.set_size_inches([15,8])
ax.plot(df["year"],df["pregnancyratetotal"],c="b")
ax.plot(df["year"],df["birthratetotal"],c="r")
ax.plot(df["year"],df["abortionratetotal"],c="Y")
ax.set_title("Comparing Abortion Rate, Pregnancy Rate and Birth Rate from 1973-2017 ")
ax.set_xlabel("Year")
ax.set_ylabel("Rate per 1000 women aged 15 - 44")
plt.xticks(rotation=90)
ax.legend(["Pregnancy Rate","Birth Rate","Abortion Rate",])
plt.show()
```

```python
merge_data = data.merge(abortion_rate,left_on=["TimeFrame","code"], right_on=["year","state"], how="inner")
```

```python
merge_data = merge_data[["Location","code","TimeFrame","Adoption_number","Foster_entry","abortion_ban","birthstotal","pregnancyratetotal","birthratetotal","abortionratetotal"]]
```

```python
merge_data['adoptions_per_1000'] = round(merge_data['Adoption_number'].dropna().astype('int')/merge_data['birthstotal'].astype('int')*1000,2)
```

# Data Visualization

```python
merge_data["TimeFrame"] = merge_data["TimeFrame"].astype("str")
```

```python
#Spliting the states based on the restrictiveness of the abortion ban based on the Time Magazine Article
from matplotlib import dates
plt.figure(figsize=(15, 8))
sns.set_theme(style="darkgrid")
ax = sns.lineplot(x="TimeFrame",y="abortionratetotal", data=merge_data.sort_values("TimeFrame"), 
                  hue='abortion_ban', marker = "o",palette=["blue","green","hotpink","orange","red"])

ax.set_xlabel("year", fontsize = 14, weight='bold')
# set y-axis label
ax.set_ylabel("Abortion rate per 1000 birth ",
              fontsize=14, weight='bold')
ax.set_title("Abortion rate based on Restrictiveness of the abortion ban ", weight='bold').set_fontsize('18')
ax.legend(["banned or mostly banned","will be banned imminently","likely to ban","legal for now","legal and likely to be protected"])
plt.show()
```

```python
#Foster care entry rate for age < 1 across All the states and D.C by year
from matplotlib import dates
plt.figure(figsize=(15, 8))
sns.set_theme(style="darkgrid")
ax = sns.lineplot(x="TimeFrame",y="fostercare_rate", data=merge_data.sort_values("TimeFrame"), hue='political_ideology', marker = "o",palette=["blue",'red','purple'])

ax.set_xlabel("year", fontsize = 14, weight='bold')
# set y-axis label
ax.set_ylabel("Foster care rate per 1000 birth ",
              fontsize=14, weight='bold')
ax.set_title("Foster care entry rate for age < 1 across All the states and D.C by year ", weight='bold').set_fontsize('18')
plt.show()
```

```python
# Comparing rates of Abortion, Adoption and Foster Care Entry
from matplotlib import dates
plt.figure(figsize=(15, 8))
sns.set_theme(style="darkgrid")
ax = sns.lineplot(x="TimeFrame",y="fostercare_rate", data=merge_data.sort_values("TimeFrame")) 
ax = sns.lineplot(x="TimeFrame",y="adoptions_per_1000", data=merge_data.sort_values("TimeFrame"))
ax = sns.lineplot(x="TimeFrame",y="abortionratetotal", data=merge_data.sort_values("TimeFrame"))
ax.lines[0].set_linestyle("--")
ax.lines[1].set_linestyle("--")
ax.lines[2].set_linestyle("--")

ax.set_xlabel("year", fontsize = 14, weight='bold')
# set y-axis label
ax.set_ylabel("Rate per 1000 births ",
              fontsize=14, weight='bold')
ax.set_title("Rates of Abortion, Adoption and Foster Care Entry", weight='bold').set_fontsize('18')
ax.legend(["Foster Care","Adoption","Abortion"])
plt.show()
```
