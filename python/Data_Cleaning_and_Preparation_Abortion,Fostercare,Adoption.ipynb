```python
#Adoption data from https://datacenter.kidscount.org/ 
adoption_rate = pd.read_excel("/content/drive/MyDrive/DS4AW Capstone Project-20220911T125049Z-001.zip (Unzipped Files)/DS4AW Capstone Project/All Data/Untitled folder/Children in child welfare system who have been adopted by age group (4).xlsx")
```

```python
#Similar filtering as above
adoption_rate = adoption_rate[(adoption_rate["Age group"]=="<1") &
                                                (adoption_rate["DataFormat"]=="Number") & ((adoption_rate["LocationType"] == 'State') | (adoption_rate["LocationType"] == 'City'))
                                                ]
```

```python
adoption_rate = adoption_rate[["Location","TimeFrame","Data"]]
```

```python
#Merging foster care and adoption data on location and year
foster_entry_adoption = adoption_rate.merge(foster_care_entry_rate, how="inner", on=("Location","TimeFrame"))
```

```python
foster_entry_adoption.rename(columns = {"Data_x":"Adoption_number","Data_y":"Foster_entry"}, inplace=True)
```

```python
foster_entry_adoption = foster_entry_adoption[["Location","TimeFrame","Adoption_number","Foster_entry"]]
```

```python
code = {'Alabama': 'AL','Alaska': 'AK','Arizona': 'AZ','Arkansas': 'AR','California': 'CA','Colorado': 'CO','Connecticut': 'CT','Delaware': 'DE',
        'District of Columbia': 'DC','Florida': 'FL','Georgia': 'GA','Hawaii': 'HI','Idaho': 'ID','Illinois': 'IL','Indiana': 'IN','Iowa': 'IA',
        'Kansas': 'KS','Kentucky': 'KY','Louisiana': 'LA','Maine': 'ME','Maryland': 'MD','Massachusetts': 'MA','Michigan': 'MI','Minnesota': 'MN',
        'Mississippi': 'MS','Missouri': 'MO','Montana': 'MT','Nebraska': 'NE','Nevada': 'NV','New Hampshire': 'NH','New Jersey': 'NJ','New Mexico': 'NM',
        'New York': 'NY','North Carolina': 'NC','North Dakota': 'ND','Ohio': 'OH','Oklahoma': 'OK','Oregon': 'OR','Pennsylvania': 'PA','Rhode Island': 'RI',
        'South Carolina': 'SC','South Dakota': 'SD','Tennessee': 'TN','Texas': 'TX','Utah': 'UT','Vermont': 'VT','Virginia': 'VA','Washington': 'WA',
        'West Virginia': 'WV','Wisconsin': 'WI','Wyoming': 'WY'}
```

```python
#Adding code of the states for merging purposes 
foster_entry_adoption["code"] = foster_entry_adoption["Location"].map(code)
```

```python
foster_entry_adoption
```

```python
#grouping states based on severity of adoption ban 1-5 scale, that data is appended 1 - banned or mostly banned, 2 - will be banned imminently, 3 - likely to ban, 4 - legal for now, 5 - legal and likely to be protected

restrictive_laws = pd.read_excel("/content/drive/MyDrive/DS4AW Capstone Project-20220911T125049Z-001.zip (Unzipped Files)/DS4AW Capstone Project/All Data/Untitled folder/abortion_val.xlsx")
```

```python
restrictive_laws["code"] = restrictive_laws["state"].map(code)
```

```python
#Merging
df = foster_entry_adoption.merge(restrictive_laws, how="inner", on="code")
```

```python
df.drop(["state"],inplace=True, axis=1)
```

```python
#Imuputing Missing Values in adoption
df["Adoption_number"].replace("N.R.",np.nan, inplace=True)
```

```python
from sklearn.impute import SimpleImputer
```

```python
#Using MODE to replace missing values 
imp_mean = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

```python
df["Adoption_number"] = imp_mean.fit_transform(df[["Adoption_number"]])
```

```python
df["Adoption_number"]
```

```python
df["Adoption_number"] = df["Adoption_number"].astype("int")
```

```python
df["Foster_entry"] = df["Foster_entry"].astype("int")
```

```python
#Downloadin Cleaned Datasets
df.to_csv("Abor_Law_Adop.csv")
```
